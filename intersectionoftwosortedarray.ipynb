# intersection of two sorted array

# naive approach

```python
def intersection(a,b,m,n):
    for i in range(m):
        if i>0 and a[i-1]==a[i]:
            continue
        for j in range(n):
            if a[i]==b[j]:
                print(a[i],end=" ")
                break

a=[2,20,20,40,60]
b=[10,20,20,20]
m=len(a)
n=len(b)
intersection(a,b,m,n)
```

# Efficient Solution

```python
def intersection(a,b,m,n):
    i=0
    j=0
    while i<m and j<n:
        if i>0 and a[i-1]==a[i]:
            i+=1
            continue
        if a[i]<b[j]:
            i+=1
        elif b[j]<a[i]:
            j+=1
        else:
            print(a[i],end=" ")
            i+=1
            j+=1

a=[2,20,20,40,60]
b=[10,20,20,20]
m=len(a)
n=len(b)
intersection(a,b,m,n)


            
        
    
```
