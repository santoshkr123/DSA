# count invrsions in an array

```python
def incCount(arr):
    n=len(arr)
    res=0
    for i in range(n-1):
        for j in range(i+1,n):
            if arr[i]>arr[j]:
                res+=1
    return res

arr=[2,4,1,3,5]
incCount(arr)
    


    

    
                
```

# efficient Solution

```python
def countMerge(arr,l,m,r):
    left=arr[l:m+1]
    right=arr[m+1:r+1]
    res,i,j,k=0,0,0,l
    while i<len(left) and j<len(right):
        if left[i]<=right[j]:
            arr[k]=left[i]
            i+=1
        else:
            arr[k]=right[j]
            j+=1
            res+=(len(left)-i)
        k+=1
        while i <len(left):
            arr[k]=left[i]
            i+=1
            k+=1
        while j<len(right):
            arr[k]=right[j]
            j+=1
            k+=1
        return res    

def countInversions(arr, l, r):
    res = 0
    if l < r:
        m = (l + r) // 2
        res += countInversions(arr, l, m)
        res += countInversions(arr, m + 1, r)
        res += countMerge(arr, l, m, r)
    return res

arr = [2, 4, 1, 3, 5]
print("Inversion count:", countInversions(arr, 0, len(arr)-1))

        
        


            
```
